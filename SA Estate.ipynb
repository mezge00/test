```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
header={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'}
pages=np.arange(1,51,1)
urls=[]
for page in pages:
    url="https://www.realtor.com/realestateandhomes-search/San-Francisco_CA/pg-"+str(page)
    urls.append(url)
    

```

```python
urls
```

```python
soups=[]
for page in urls:
    response=requests.get(page,headers=header)
    soup=BeautifulSoup(response.content,'lxml')
    soups.append(soup.select('.BasePropertyCard_propertyCard__JXqlE'))
```

```python
alist = []
for sublist in soups:
    for item in sublist:
        alist.append(item)
len(alist)
```

```python
allprice=[]
for item in alist:
    try:
        priceList=soup.find_all("div","card-price")
        allprice.append(priceList)
       # print(item.select('[Pricestyles__StyledPrice-rui__btk3ge-0 bvgLFe card-price=card-price]')[0].get_text())
    except Exception as e:
        print(" ")
        
    
```

```python
fallprice = []
for sublist in allprice:
    for item in sublist:
        fallprice.append(item)
```

```python
data={'price':[]}
for j in range(len(fallprice)):
    #data['item'].append(itemsList[j].text)
    data['price'].append(fallprice[j].text)
```

```python
df=pd.DataFrame(data)
```
